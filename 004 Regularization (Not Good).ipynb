```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Import My Pre-defined Linear Model
from t_model import LinearRegressionModel
from t_model_regu import LinearRegressionModelWithRegularization
from t_preprocessing import polynomial_features
from t_visualization import plot_polynomial_curve
```

# Gather the Data

```python
np.random.seed(27)

# Define expected Function
m = 100
X = 6 * np.random.rand(m, 1) - 2
y = 1.5 * np.square(X) + 1*X + 2 + np.random.randn(m, 1)

# Plot the Data
plt.figure(figsize=(16, 6))
plt.plot(X, y, "b.")
plt.xlabel("$x_1$")
plt.ylabel("$y$", rotation=0)
plt.show()
```

```python
model_non_regu = LinearRegressionModel(polynomial_features(X, degree=10).shape[1], learning_rate=2e-11, n_estimators=1000)
```

```python
# Set the n_estimators to 400 because it's the optimal cost, though it still overfit there.
model_regu = LinearRegressionModelWithRegularization(polynomial_features(X, degree=10).shape[1], learning_rate=2e-11, n_estimators=400, lambd=0.8)
```

## Train Models

```python
# This model fitting run with 50 * 1000 = 50.000 iterations
for i in range(50):
    model_non_regu.fit(polynomial_features(X, degree=10), y)
```

```python
# And this regularized model run with just 400 iterations :)
model_regu.fit(polynomial_features(X, degree=10), y)
```

```python
# Plot These Model Before tranining
# Plot the Data
plt.figure(figsize=(18, 10))
plt.plot(X, y, "b.")
plt.xlabel("$x_1$")
plt.ylabel("$y$", rotation=0)

# Plot the Model Curves
plot_polynomial_curve(model_non_regu, 'r', 'Without Regularization')
plot_polynomial_curve(model_regu, 'g', 'With Regularization')

plt.legend(fontsize=13, loc='upper left')
plt.ylim([-10, 50])
plt.show()
```
